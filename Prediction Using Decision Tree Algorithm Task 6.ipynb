# Prediction Using Decision Tree Algorithm Task 6

### First I will import my libraries

```python
import pandas as pd
import matplotlib.pyplot as plt 
```

### Reading my data

```python
df = pd.read_csv("Iris.csv")
target = df.Species
features = df.drop({"Species", "Id"}, axis =1)
```

### Preparing my data and split it

```python
from sklearn.model_selection import train_test_split
x_train, x_val, y_train, y_val = train_test_split(features, target, test_size = 0.2)
```

### Training my mode

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier().fit(x_train,y_train)
```

### Evaluate my model

```python
from sklearn.metrics import accuracy_score
pred = model.predict(x_val)
print("The Accuracy Score = {}".format(accuracy_score(y_val, pred)))
```

### Let us visualize the Decision Tree to understand it better

```python
from sklearn.tree import plot_tree, export_text
fig = plt.figure()
ax = fig.add_axes([6,6, 9,9])
plot_tree(model, feature_names = features.columns, class_names= 'Species' ,filled = True)
```
